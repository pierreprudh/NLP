```python
import nltk 
from nltk.util import ngrams
```

```python
def extract_ngrams(data, num) : 
    n_grams = ngrams(nltk.word_tokenize(data), num)
    return [ '  '.join(grams) for grams in n_grams]
```

```python
data = 'A class is a blueprint for the object. Backgammon is one of the oldest known board games.'
```

```python
for i in range(1, 7) : 
    print(f"{i}-gram : ", extract_ngrams(data,i))
```

```python
import pandas as pd 
import nltk
import numpy as np 

from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize as st 
from nltk.stem import WordNetLemmatizer as wordnet
import re
```

```python
df = pd.read_csv('./data/spam.csv', encoding = 'ISO-8859-1', usecols = ['v1', 'v2'])

corpus = []
wordnet = wordnet()
length = len(df['v2'])

df.head(10)
```

```python
for i in range(length) : 
    
    rev = re.sub('[^a-zA-Z]', ' ', df['v2'][i])
    rev = rev.lower()
    rev = rev.split()
    rev = [ wordnet.lemmatize(word) for word in rev if word not in stopwords.words('english')]
    rev = ' '.join(rev)

    corpus.append(rev)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features= 2500)
x = cv.fit_transform(corpus).toarray()

y = df['v1']
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

y_pred = model.predict(x_test)

model.score(x_test, y_test)

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
import re 

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, f1_score

sns.set_context('talk')
sns.set_color_codes()
plot_kwds = {'alpha': 0.25, 's' : 80, 'linewidths' : 0}

import warnings; warnings.simplefilter('ignore')
```

```python
data_file = './data/Comment Spam.xls'

data = pd.read_excel(data_file)
data = data[['Comment', 'Class']]
train_data = data
train_data.head()
```

```python
def process_content(content) : 
    return " ".join(re.findall("[A-Za-z]+", content.lower()))

train_data['processed_comment'] = train_data['Comment'].apply(process_content)
train_data.head(10)
```

```python
x_train, x_test, y_train, y_test = train_test_split(train_data['processed_comment'], train_data['Class'], test_size=0.2)
```

```python
model = Pipeline([( 'vect', CountVectorizer(stop_words="english")),
                  ('tfid', TfidfTransformer()),
                  ('clf', LogisticRegression()),
])
```

```python
model.fit(x_train, y_train)

predicted = model.predict(x_test)

cm = confusion_matrix(y_test, predicted)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
print('accuracy_score', accuracy_score(y_test, predicted))
print('Reporting...')
print(classification_report(y_test, predicted))
```

```python
print(cross_val_score(model, x_train, y_train, cv = 5))
print(cross_val_score(model, x_test, y_test, cv = 5))
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
import warnings
import gensim 
from gensim.models import Word2Vec

warnings.filterwarnings(action = 'ignore')
```

```python
sample = open('./data/alice_wonderland.txt')
s = sample.read()
```

```python
f = s.replace("\n", " ")
data = []

for i in sent_tokenize(f) : 
    temp = []
    
    for j in word_tokenize(i) : 
        temp.append(j.lower())

    data.append(temp)
```

```python
for i in [2, 10, 500] : 
    model1 = gensim.models.Word2Vec(data, min_count = 1, vector_size=i, window = 5)

    print("Model with vector size : ", i)
    print("Cosine similarity between 'alice' " + "and 'wonderland' - CBOW :", model1.wv.similarity('alice', 'wonderland'))
    print("Cosine similarity between 'alice' " + "and 'machines' - CBOW :", model1.wv.similarity('alice', 'machines'))
```

```python
for i in [2, 10, 500] : 
    model2 = gensim.models.Word2Vec(data, min_count = 1, vector_size=i, window = 5, sg = 1)
    
    print("Model with vector size : ", i)
    print("Cosine similarity between 'alice' " + "and 'wonderland' - SkipGram :", model2.wv.similarity('alice', 'wonderland'))
    print("Cosine similarity between 'alice' " + "and 'machines' - SkipGram :", model2.wv.similarity('alice', 'machines'))
```
